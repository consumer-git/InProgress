```python
df2=pd.merge(unilever, 
         hdfc, 
         right_index=True, 
         left_index=True,
         suffixes=('_unilever', '_hdfc')
)
```

```python
df3=pd.merge(df, 
         df2, 
         right_index=True, 
         left_index=True)
```

```python
df_final=pd.merge(df3, 
         nifty50, 
         right_index=True, 
         left_index=True)
```

## next steps - scale data 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaly = MinMaxScaler() 
```

```python
v=['Close Price_infy','Close Price_tcs','Close Price_unilever','Close Price_hdfc','Close Price_infy','close_n50']
```

```python
df_final[v] = scaly.fit_transform(df_final[v])
```

```python
new_df=df_final[v]
```

```python
new_df=pd.DataFrame(new_df)
```
